```julia
function approxcos(x)
    outval = Array(Any, size(x))
    
    ii = 0
    for aa in x
        y = 1 - aa^2/2 + aa^4/24 - aa^6/720 + aa^8/(56*720)
        ii += 1
        outval[ii] = y
    end
    
    return outval
end
```

```julia
x1 = 6rand(9) # 6 times rand()
x2 = linspace(0, 6, 200)

y1 = approxcos(x1)
y2 = cos(x2)
```

```julia
using Plots; pyplot()

plot(x1, y1, line=:scatter, legend=:false, title="Illustrating 6th order approximation to cos")

plot!(x2, y2, line=(:path, :red))
```
